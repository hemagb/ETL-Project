```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
food_data = "Resources/Food_Establishment_Inspection_Scores.csv"
rest_df = pd.read_csv(food_data)
rest_df
```

```python
rest_columns = ["Facility ID", "Restaurant Name", "Zip Code", "Inspection Date", "Score"]
rest_transformed = rest_df[rest_columns].copy()

rest_transformed = rest_transformed.rename(columns={'Facility ID': 'facility_id',
                                            'Restaurant Name': 'restaurant_name',
                                           'Zip Code': 'zip_code',
                                            'Inspection Date': 'inspect_date',
                                           'Score': 'inspect_score',
                                           })

rest_transformed.drop_duplicates("facility_id", inplace=True)
rest_transformed.set_index("facility_id", inplace=True)
rest_transformed
```

```python
connection = "root:MyNewPass@localhost/restaurant_scores_db"
engine = create_engine(f'mysql://{connection}')
```

```python
engine.table_names()
```

```python
rest_transformed.to_sql(name='restaurant_scores', con=engine, if_exists='append', index=True)
```
