```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
csv_file = 'Resources/austin-public-health-locations-1.csv'
public_health_df = pd.read_csv(csv_file)
public_health_df.head()
```

```python
clean_public_health_df = public_health_df[['Facility Name', 'Street Address', 'Zip Code']].copy()
clean_public_health_df.head()
```

```python
clean_public_health_df.dropna()
```

```python
clean_public_health_df.rename(columns={
    'Facility Name': 'facility_name',
    'Street Address': 'street_address',
    'Zip Code': 'zip_code'}, inplace=True)
clean_public_health_df.head()
```
